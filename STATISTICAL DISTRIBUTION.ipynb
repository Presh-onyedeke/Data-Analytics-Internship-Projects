```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

a = 8 #lambda

x = np.linspace(0, 20,20)
my_var = poisson (a)
fig, _axis = plt.subplots(1,1)
_axis.vlines(x, 0, my_var.pmf(x), label = 'data')
plt.show()
```

```python
import numpy as np

import matplotlib.pyplot as plt
from scipy.stats import expon
lam = 4
num = 100000
waiting_time = expon(scale = 1/lam).rvs(num)
pd.Series(waiting_time).hist(bins = 20)
plt.title ('distance between event')
plt.xlabel('waiting events')
plt.ylabel('frequency')
plt.show()
```

```python
import numpy as np

import matplotlib.pyplot as plt
from scipy.stats import gamma
lam = 3
num = 100000
waiting_time = gamma(10, scale = 1/lam).rvs(num)
pd.Series(waiting_time).hist(bins = 20)
plt.title ('distance between event')
plt.xlabel('waiting events')
plt.ylabel('frequency')
plt.show()
```

```python
import numpy as np

import matplotlib.pyplot as plt
from scipy.stats import weibull_min
lam = 4
num = 100000
waiting_time = weibull_min(c=1.4, scale = 1/lam).rvs(num)
pd.Series(waiting_time).hist(bins = 20)
plt.title ('distance between event')
plt.xlabel('waiting events')
plt.ylabel('frequency')
plt.show()
```
