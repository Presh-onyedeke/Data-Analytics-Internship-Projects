```python
import requests  #importing requests
```

```python
site_url = 'https://pureportal.coventry.ac.uk/en/organisations/school-of-economics-finance-and-accounting/publications/'
response = requests.get(site_url)
```

```python
response.status_code   #status of the site downloaded
```

```python
len(response.text)  #lenght of the characters in the page
```

```python
page_contents=response.text
```

```python
!pip install beautifulsoup4 --quiet
```

```python
from bs4 import BeautifulSoup #importing beautifulsoup
```

```python
doc=BeautifulSoup(page_contents, 'html.parser' )
```

```python
req_path=doc.find_all('a',{'rel':'ContributionToJournal'})
req_path
```

```python
publication=[]
for i in req_path:
    title = i.find_all('span')
    for tags in title:
        publication.append(tags.text)
print(publication)
    
```

```python
pub_url=[]
for a in req_path:
    pub_url.append(a['href'])
print(pub_url)
```

```python
author=doc.find_all('div',{'class':'rendering rendering_researchoutput rendering_researchoutput_portal-short rendering_contributiontojournal rendering_portal-short rendering_contributiontojournal_portal-short'},'span')
print(author)
```

```python
main_auth=[]
for i in author:
    auth=i.find_all('span',{'class':""})[1:-1]
    print(auth)
    main_auth.append(auth)
        
```

```python
!pip install pandas --quiet
```

```python
import pandas as pd
```

```python
web_data_dict={
    'Publication':publication,
    'Authors':main_auth,
    'Publication-link':pub_url
}
```

```python
web_data_df = pd.DataFrame(web_data_dict)
```
